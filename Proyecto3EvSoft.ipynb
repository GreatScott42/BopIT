```python
#t test
from scipy import stats
#

aesthetics_cols = ['AE1','AE2','AE3','AE4']
play_engrossment_cols = ['PE1','PE2','PE3','PE4','PE5','PE6','PE7']
enjoyment_cols = ['EN1','EN2','EN3','EN4','EN5','EN6']
personal_gratification_cols = ['PG1','PG2','PG3','PG4']
creative_freedom_cols = ['CF1','CF2','CF3','CF4']
social_connectivity_cols = ['SC1','SC2','SC3','SC4','SC5']
narrative_cols = ['NA1','NA2','NA3','NA4']
usability_cols = ['US1','US2','US3','US4','US5','US6','US7']
visual_aesthetics_cols = ['VA1','VA2','VA3','VA4']

datav1['Aesthetics'] = datav1[aesthetics_cols].mean(axis=1)
datav1['Play Engrossment'] = datav1[play_engrossment_cols].mean(axis=1)
datav1['Enjoyment'] = datav1[enjoyment_cols].mean(axis=1)
datav1['Personal Gratification'] = datav1[personal_gratification_cols].mean(axis=1)
datav1['Creative Freedom'] = datav1[creative_freedom_cols].mean(axis=1)
datav1['Social Connectivity'] = datav1[social_connectivity_cols].mean(axis=1)
datav1['Narrative'] = datav1[narrative_cols].mean(axis=1)
datav1['Usability/Playability'] = datav1[usability_cols].mean(axis=1)
datav1['Visual Aesthetics'] = datav1[visual_aesthetics_cols].mean(axis=1)

datav2['Aesthetics'] = datav2[aesthetics_cols].mean(axis=1)
datav2['Play Engrossment'] = datav2[play_engrossment_cols].mean(axis=1)
datav2['Enjoyment'] = datav2[enjoyment_cols].mean(axis=1)
datav2['Personal Gratification'] = datav2[personal_gratification_cols].mean(axis=1)
datav2['Creative Freedom'] = datav2[creative_freedom_cols].mean(axis=1)
datav2['Social Connectivity'] = datav2[social_connectivity_cols].mean(axis=1)
datav2['Narrative'] = datav2[narrative_cols].mean(axis=1)
datav2['Usability/Playability'] = datav2[usability_cols].mean(axis=1)
datav2['Visual Aesthetics'] = datav2[visual_aesthetics_cols].mean(axis=1)


t_aest, p_aest = stats.ttest_rel(datav1['Aesthetics'], datav2['Aesthetics'])
t_playEngross, p_playEngross = stats.ttest_rel(datav1['Play Engrossment'], datav2['Play Engrossment'])
t_enjoy, p_enjoy = stats.ttest_rel(datav1['Enjoyment'], datav2['Enjoyment'])
t_personalGrat, p_personalGrat = stats.ttest_rel(datav1['Personal Gratification'], datav2['Personal Gratification'])
t_creativeFree, p_creativeFree = stats.ttest_rel(datav1['Creative Freedom'], datav2['Creative Freedom'])
t_socialConnect, p_socialConnect = stats.ttest_rel(datav1['Social Connectivity'], datav2['Social Connectivity'])
t_narrative, p_narrative = stats.ttest_rel(datav1['Narrative'], datav2['Narrative'])
t_usability, p_usability = stats.ttest_rel(datav1['Usability/Playability'], datav2['Usability/Playability'])
t_visualAest, p_visualAest = stats.ttest_rel(datav1['Visual Aesthetics'], datav2['Visual Aesthetics'])

final = pd.DataFrame({
    'Subescala': ['Aesthetics', 'Play Engrossment', 'Enjoyment',
                  'Personal Gratification', 'Creative Freedom',
                  'Social Connectivity', 'Narrative',
                  'Usability/Playability', 'Visual Aesthetics'],
    'p-valor': [p_aest, p_playEngross, p_enjoy,
                   p_personalGrat, p_creativeFree,
                   p_socialConnect, p_narrative,
                   p_usability, p_visualAest]
})
final
```
